```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.histplot(iris)
```

```python
sns.histplot(iris, x='sepal_length')
```

```python
sns.histplot(iris, x='sepal_length', bins=15, kde=True ) #density
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris, x='sepal_width', bins=15, kde=True)
plt.savefig('sepal_width_hist.png', dpi=300)
```

```python
sns.histplot(iris, x='petal_length')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris, x='petal_width', bins=5, kde=True)
plt.savefig('petal_width_hist.png', dpi=300)
```

```python
sns.histplot(iris, x='petal_length', bins=15, kde=True)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris, x='petal_length', bins=15, kde=True)
plt.savefig('petal_length_hist.png', dpi=300)
```

```python
sns.set_palette('rainbow')
fig, ax = plt.subplots(figsize = (20,5), ncols= 4)
sns.histplot(iris, x='sepal_width', bins=15, kde=True , ax = ax[0] , color='red')
sns.histplot(iris, x='petal_width', bins=15, kde=True, ax = ax[1])
sns.histplot(iris, x='sepal_length', bins=15, kde=True, ax = ax[2])
sns.histplot(iris, x='petal_length', bins=15, kde=True, ax = ax[3] )
```

```python
sns.set_palette('rainbow')
fig, ax = plt.subplots(figsize = (20,8), ncols= 2, nrows=2)
sns.histplot(iris, x='sepal_width', bins=15, kde=True , ax = ax[0,0] , color='red')
sns.histplot(iris, x='petal_width', bins=15, kde=True, ax = ax[1,0])
sns.histplot(iris, x='sepal_length', bins=15, kde=True, ax = ax[0,1])
sns.histplot(iris, x='petal_length', bins=15, kde=True, ax = ax[1,1 ],color='orange')
```
